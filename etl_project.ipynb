```python
# Importing variables
import pandas as pd
from pandas.io import sql
from sqlalchemy import create_engine, inspect
```

## Extracting CSV into DataFrame

```python
datafiniti_file = "./Resources/Datafiniti_Pizza.csv"
datafiniti_df = pd.read_csv(datafiniti_file)
datafiniti_df.head()
```

```python
pizza_dataset_file = "./Resources/Pizza_Dataset.csv"
pizza_df = pd.read_csv(pizza_dataset_file)
pizza_df.head()
```

## Transforming Datafiniti_File

```python
# Creating a filtered dataframe from specific columns

datafiniti_cols = ["id", "dateUpdated", "address", "categories", "city", "country",
                   "menus.description", "menus.name", "name", "postalCode", "priceRangeMin",
                   "priceRangeMax", "province"]

datafiniti_transformed = datafiniti_df[datafiniti_cols].copy()

# Renaming the column headers

datafiniti_transformed = datafiniti_transformed.rename(columns={"id": "SiteID", "dateUpdated": "DateUpdated",
                                                                "address": "Address", "categories": "Categories",
                                                                "city":"City","country":"Country", 
                                                                "menus.description": "Menu_Description",
                                                                "menus.name" : "Menu_Name", "name": "Name", 
                                                                "postalCode" : "Postal_Code", 
                                                                "priceRangeMin": "Price_Range_Min", 
                                                                "priceRangeMax": "Price_Range_Max",
                                                                "province" : "Province"})
# Setting the Index
datafiniti_transformed.set_index

# Reviewing 
datafiniti_transformed.head()
```

## Transforming Pizza_Dataset_File

```python
# Creating a filtered dataframe from specific columns

pizza_dataset_cols = ["id", "keys", "latitude", "longitude"]

pizza_dataset_transformed = pizza_df[pizza_dataset_cols].copy()

# Renaming the column headers

pizza_dataset_transformed = pizza_dataset_transformed.rename(columns={"id": "SiteID", "keys":"Key", 
                                                                      "latitude": "Latitude", 
                                                                      "longitude":"Longitude"})

# Setting the Index
pizza_dataset_transformed.set_index

# Reviewing
pizza_dataset_transformed.head()
```

## Loading DataFrames into database

```python
pizza_dataset_transformed.to_sql(name='Pizza_Dataset', con=engine, if_exists = "replace")
datafiniti_transformed.to_sql(name='Datafiniti_Pizza', con=engine, if_exists = "replace" )
```
